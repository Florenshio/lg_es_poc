# 학습데이터 구축

```python
# Slplit images
import os
import shutil
import datetime
import random

def makedirs(path):
    if not os.path.exists(path):
        os.makedirs(path)

# 오늘 날짜
today = datetime.datetime.today().strftime('%Y%m%d')[2:]

''' =================== 해당 부분의 경로만 레이블링한 이미지와 텍스트 파일이 있는 경로로 변경 ====================================================='''

# 이미지 및 텍스트 파일 
files = os.listdir('/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/00. data_final')

'''ver0.5 - 2023.10.21'''
# 이미지 및 텍스트 파일 경로 - 레이블링한 이미지와 텍스트가 있는 경로
all_files_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/00. data_final'
'''============================================================================================================================================='''

# 이동할 경로
train_images_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/train/images/'
train_labels_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/train/labels/'
val_images_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/val/images/'
val_labels_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/val/labels/'

# train, val, test 비율 설정
train_ratio = 0.8
val_ratio = 0.2
# test_ratio=0.1

# 디렉토리 생성
makedirs(train_images_directory)
makedirs(train_labels_directory)
makedirs(val_images_directory)
makedirs(val_labels_directory)

# 파일 복사
for file in files:
    if file.lower().endswith(('.png','.jpg','.jpeg')):
        # 이미지 파일일 경우
        image_path = os.path.join(all_files_directory,file)
        txt_file = file.replace(os.path.splitext(file)[1],'.txt')
        txt_path = os.path.join(all_files_directory, txt_file)
        
        if os.path.exists(txt_path):
            # 텍스트 파일이 존재하는 경우
            rand = random.random()
            if rand < train_ratio:
                destination_image_path = os.path.join(train_images_directory,file)
                destination_txt_path = os.path.join(train_labels_directory,txt_file)
            elif rand < train_ratio + val_ratio:
                destination_image_path = os.path.join(val_images_directory,file)
                destination_txt_path = os.path.join(val_labels_directory,txt_file)
            # else:
            #     destination_image_path = os.path.join(test_directory,file)
            #     destination_txt_path = os.path.join(test_directory,txt_file)
                
            shutil.copyfile(image_path,destination_image_path)
            shutil.copyfile(txt_path,destination_txt_path)

# # yaml 파일 수정
# with open('/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/yaml/train.yaml','r') as y:
#     yaml = y.readlines()
    
# # 이미지 경로 수정
# yaml[1] = yaml[1][:-8] + today +'/\\n'

# with open ('/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/yaml/train.yaml','w') as f:
#     f.write(''.join(yaml))
```

# Pre-trained Model 로드

```python
# Load Model
import os
from ultralytics import YOLO

os.environ['CUDA_DEVICE_ORDER'] = 'PCI_BUS_ID'
os.environ['CUDA_VISIBLE_DEVICES'] = '1'

model = YOLO('/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/03. 파인튜닝 모델/yolov8l.pt')
```

# 학습 시작

```python
# Start Train
from setproctitle import *
from datetime import datetime

now = datetime.now()
setproctitle(f'YOLOv8_Train_{now.date()}')
os.environ['OMP_NUM_THREADS'] = '16'

# export OMP_NUM_THREADS = 3
model.train(data = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/yaml/train.yaml', epochs = 1000, imgsz=975, patience = 50,save_period = 100, batch = 4,
           scale=0.3)
```
