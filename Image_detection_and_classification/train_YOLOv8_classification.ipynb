```python
def reset_directory(path):
    # 해당 디렉토리 모든 내용 삭제
    files = glob.glob(path+'*')
    for file in files:
        os.remove(file)
        
def rm_dir(path):
    shutil.rmtree(path)

def makedirs(path):
    if not os.path.exists(path):
        os.makedirs(path)
        
# 오늘 날짜
today = datetime.datetime.today().strftime('%Y%m%d')[2:]

# train 디렉토리 초기화
#rm_dir(f'/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/{today}/')

# 클래스별 이미지 리스트
cell = glob.glob('/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/231120_cls_임시보관소/train/cell/*.png')
assy = glob.glob('/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/231120_cls_임시보관소/train/assy/*.png')
roll = glob.glob('/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/231120_cls_임시보관소/train/roll/*.png')
rode = glob.glob('/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/231120_cls_임시보관소/train/rode/*.png')
etc = glob.glob('/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/231120_cls_임시보관소/train/etc/*.png')
#circuit = glob.glob('/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/231120_cls_임시보관소/train/circuit/*.png')

print(f'cell size : {len(cell)}')
print(f'assy size : {len(assy)}')
print(f'roll size : {len(roll)}')
print(f'rode size : {len(rode)}')
print(f'etc size : {len(etc)}')
#print(f'circuit size : {len(circuit)}')

# train
cell_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/train/cell/'
assy_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/train/assy/'
roll_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/train/roll/'
rode_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/train/rode/'
etc_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/train/etc/'
#circuit_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/train/circuit/'

# val
val_cell_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/val/cell/'
val_assy_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/val/assy/'
val_roll_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/val/roll/'
val_rode_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/val/rode/'
val_etc_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/val/etc/'
#val_circuit_directory = '/data/LG_ES_POC/002. PoC_개발/09. 핵심이미지 식별/another_workspace/03. 파인튜닝 모델/train/'+ today +'/val/circuit/'

# 디렉토리 생성
makedirs(cell_directory)
makedirs(assy_directory)
makedirs(roll_directory)
makedirs(rode_directory)
makedirs(etc_directory)
#makedirs(circuit_directory)
makedirs(val_cell_directory)
makedirs(val_assy_directory)
makedirs(val_roll_directory)
makedirs(val_rode_directory)
makedirs(val_etc_directory)
#makedirs(val_circuit_directory)

# train, val, test 비율 설정
train_ratio = 0.9
val_ratio = 0.1

for n,category in enumerate([cell,assy,roll,rode,etc]):
        
    # 파일 복사
    for file in category:
                
        name = file.split('/')[-1]
        rand = random.random()
        if rand < train_ratio:
            if n == 0:
                shutil.copyfile(file,cell_directory+name)
            if n == 1:
                shutil.copyfile(file,assy_directory+name)
            if n == 2:
                shutil.copyfile(file,roll_directory+name)
            if n == 3:
                shutil.copyfile(file,rode_directory+name)
            if n == 4:
                shutil.copyfile(file,etc_directory+name)
            # if n == 5:
            #     shutil.copyfile(file,circuit_directory+name)
            
        elif rand < train_ratio + val_ratio:
            if n == 0:
                shutil.copyfile(file,val_cell_directory+name)
            if n == 1:
                shutil.copyfile(file,val_assy_directory+name)
            if n == 2:
                shutil.copyfile(file,val_roll_directory+name)
            if n == 3:
                shutil.copyfile(file,val_rode_directory+name)
            if n == 4:
                shutil.copyfile(file,val_etc_directory+name)
            # if n == 5:
            #     shutil.copyfile(file,val_circuit_directory+name)
```
